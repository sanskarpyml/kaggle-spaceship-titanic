Data preprocessing and Data cleaning

```python
# Filling  missing categorical label like 'HomePlanet', 'Destination', 'CryoSleep', 'VIP' with mode
for col in ['HomePlanet', 'Destination', 'CryoSleep', 'VIP']:
    train[col].fillna(train[col].mode()[0], inplace=True)
null_counts = train.isnull().sum()
print(null_counts)
```

```python
# Fill missing 'Age' and spending columns with median
for col in ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']:
    train[col].fillna(train[col].median(), inplace=True)
null_counts = train.isnull().sum()
print(null_counts)
```

Feature Engineering

```python
# Splitting Cabin into Deck / Num / Side
train[['Deck', 'Cabin_num', 'Side']] = train['Cabin'].str.split('/', expand=True)
train.drop('Cabin', axis=1, inplace=True)
```

```python
train.head()
```

```python
# Filling missing values in new features
for col in ['Deck', 'Cabin_num', 'Side']:
    train[col].fillna(train[col].mode()[0], inplace=True)
null_counts = train.isnull().sum()
print(null_counts)
```

```python
# Creating TotalSpend feature which will include all the bills like RoomService FoodCourt and etc
train['TotalSpend'] = train['RoomService'] + train['FoodCourt'] + \
                      train['ShoppingMall'] + train['Spa'] + train['VRDeck']
#As we created new feature total spend we can drop down individual bills
train.drop(['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck'], axis=1, inplace=True)
```

```python
train.head()
```

```python
# Listing of categorical features
categorical_features = ['HomePlanet', 'Destination', 'CryoSleep', 'VIP', 'Deck', 'Side', 'Cabin_num']
# Encoding categorical features using Label Encoding
le = LabelEncoder()
for col in categorical_features:
    train[col] = le.fit_transform(train[col])
train.head()
```

```python
#Encoding target variable
train['Transported'] = train['Transported'].map({True: 1, False: 0})
train.head()
```

```python
#Features and target
X = train.drop(['PassengerId', 'Transported', 'Name'], axis=1)
Y = train['Transported']
```

traing model

```python
#Building a  model ()
import lightgbm as lgb
model = lgb.LGBMClassifier(
    n_estimators=500,
    learning_rate=0.03,
    max_depth=7,
    random_state=42,
    subsample=0.8,
    colsample_bytree=0.8)
model.fit(X, Y)
y_pred = model.predict(X)
accuracy = accuracy_score(Y, y_pred)
print(f"Validation Accuracy: {accuracy:.4f}")
```
